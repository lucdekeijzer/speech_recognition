```python
import torchaudio
import os

def get_wav_duration(filepath):
    waveform, sample_rate = torchaudio.load(filepath)
    duration = waveform.size(1) / sample_rate  # waveform size[1] gives the number of samples
    return duration

def average_wav_length(directory):
    total_duration = 0
    num_files = 0
    
    for filename in os.listdir(directory):
        if filename.endswith(".wav"):
            filepath = os.path.join(directory, filename)
            duration = get_wav_duration(filepath)
            total_duration += duration
            num_files += 1
    
    if num_files == 0:
        return 0
    
    return total_duration / num_files

# Example usage
directory = 'D:/Universiteit Umea/HT24/Student Conference/speech_recognition/7946870/20221103_gos/20221103_gos/clips-clean/'
average_length = average_wav_length(directory)
print(f"The average length of .wav files is {average_length:.2f} seconds")
```

```python
import wave
import contextlib
import os

def get_wav_duration(filepath):
    with contextlib.closing(wave.open(filepath, 'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        return duration

def average_wav_length(directory):
    total_duration = 0
    num_files = 0
    
    for filename in os.listdir(directory):
        if filename.endswith(".wav"):
            filepath = os.path.join(directory, filename)
            duration = get_wav_duration(filepath)
            total_duration += duration
            num_files += 1
    
    if num_files == 0:
        return 0
    
    return total_duration / num_files

# Example usage
directory = 'D:/Universiteit Umea/HT24/Student Conference/speech_recognition/7946870/20221103_gos/20221103_gos/clips-clean/'
average_length = average_wav_length(directory)
print(f"The average length of .wav files is {average_length:.2f} seconds")
```
